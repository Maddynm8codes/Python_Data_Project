```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_India = df[df['job_country'] == 'India'].copy()
df_India
```

```python
df_India['job_posted_month'] = df_India['job_posted_date'].dt.strftime('%B')
```

```python
#Pivoting with job month along the index and job titles in the column

df_pivot = df_India.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
```

```python
df_pivot.reset_index(inplace=True)
```

```python
df_pivot['month_no'] = pd.to_datetime(df_pivot['job_posted_month'], format='%B').dt.month
```

```python
df_pivot.sort_values('month_no',inplace=True) 
df_pivot.set_index('job_posted_month', inplace=True)
df_pivot.drop(columns='month_no', inplace=True)
```

```python
df_pivot
```

```python
df_pivot.plot(kind='line')
```

```python
top_3 = df_India['job_title_short'].value_counts().head(3)
top_3 = top_3.index.to_list()
```

```python
# Create the line plot
df_pivot[top_3].plot(kind='line')

# Set the x-axis label, y-axis label, and title
plt.xlabel('Month')
plt.ylabel('Job Count')
plt.title('Monthly Job Postings for Data Jobs in India (Top 3)')

# Set x-ticks to show every month
plt.xticks(range(len(df_pivot.index)), df_pivot.index, rotation=45)  # Adjust rotation as needed

# Display the legend
plt.legend()

# Show the plot
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
df_software = pd.read_csv('https://lukeb.co/software_csv', index_col='job_posted_month')
```

```python
df_merged = df_pivot.merge(df_software, on='job_posted_month')
```

```python
top_5 = (df_merged
         .sum()
         .sort_values(ascending=False)
         .head()
         .index
         .to_list()
)
```
